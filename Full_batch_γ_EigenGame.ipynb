```python
import numpy as np


def generate_random_symmetric(n):
    a = np.random.uniform(0, 1, (n, n))
    return a + a.T - np.diag(a.diagonal())


def generate_random_symmetric_positive_definite_matrix(n):
    a = np.random.uniform(0, 1, (n, n))
    return a @ a.T

np.random.seed(0)

n = 4

A = generate_random_symmetric(n=n)
B = generate_random_symmetric_positive_definite_matrix(n=n)

A, B
```

```python
import scipy.linalg

eigenvalues, eigenvectors = scipy.linalg.eigh(A, B)
print("eigenvalues", eigenvalues)
print("eigenvectors", eigenvectors)
```

```python
# Full-batch γ-EigenGame
```

```python
T = int(1e5)
k = 2

vs = np.random.normal(size=(k, n,))
vs = vs / np.linalg.norm(vs, axis=1)[:, None]

eta = 1e-3
for t in range(T):
  for i in range(k):
    v_i =  vs[i, :]
    B_v_i = B @ v_i
    A_v_i = A @ v_i
    v_i_T_B_v_i = v_i.T @ B_v_i
    rewards = v_i_T_B_v_i * A_v_i - (v_i.T @ A_v_i) * B_v_i
    penalties = 0
    for j in range(i):
      v_j = vs[j, :]
      y_j = v_j / np.sqrt(v_j.T @ B @ v_j)
      B_y_j = B @ y_j
      penalties += (v_i.T @ A @ y_j) * (v_i_T_B_v_i * B_y_j - (v_i @ B_y_j) * B_v_i)
    if (t % 100) == 0:
      print("t", t, "i", i, "r", rewards, "penalties", penalties)
    d_i = rewards - penalties
    new_v_i = v_i - eta * d_i
    new_v_i /=  np.linalg.norm(new_v_i)
    vs[i, :] = new_v_i

lmds = np.empty(k,)
for i in range(k):
  v_i =  vs[i, :]
  B_v_i = B @ v_i
  A_v_i = A @ v_i
  lmds[i] = (v_i.T @ A_v_i) / (v_i.T @ B_v_i)

print("eigenvalues", lmds)
print("eigenvectors", vs)
```
